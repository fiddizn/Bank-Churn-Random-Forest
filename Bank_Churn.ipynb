```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df = pd.read_csv('/content/train.csv')
df.head()
```

```python
df.info()
```

```python
df.Exited.value_counts()
```

```python
# Mengambil sampel acak untuk kategori Exited dengan nilai 0
df_exited_0_sample = df[df['Exited'] == 0].sample(n=50000, random_state=354)

# Menggabungkan data yang memiliki nilai Exited 1 dengan data sampel Exited 0
df = pd.concat([df[df['Exited'] == 1], df_exited_0_sample])
```

```python
df.Exited.value_counts()
```

```python
features = ['CreditScore', 'Age', 'Balance', 'NumOfProducts', 'IsActiveMember']
```

```python
# Memisahkan data menjadi fitur (X) dan target (y)
X = df[features]
y = df['Exited']
```

```python
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Memisahkan data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Membuat model Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Melatih model
rf_classifier.fit(X_train, y_train)

# Membuat prediksi pada data testing
y_pred = rf_classifier.predict(X_test)
```

```python
# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
# Menampilkan hasil evaluasi
print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```
